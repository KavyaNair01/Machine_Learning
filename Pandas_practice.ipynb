### Creating new columns in a dataframe

```python
df.columns
```

create a column by dividing video views by subscribers and name it as views per subscriber and print min, max , mean and no of null values of the new column


```python
df['Views_per_subscriber'] = df['video views']/df['subscribers']
df['Views_per_subscriber'].min(), df['Views_per_subscriber'].max(), df['Views_per_subscriber'].mean(), df['Views_per_subscriber'].isnull().sum()
```

Create a column with create year, month and date as one column

```python
df['create_dt'] = df['created_date'].astype(str) + '-' +  df['created_month'] + '-' + df['created_year'].astype(int, errors='ignore').astype(str)
df['create_dt']
```

each country lo what is the average of video views per subscriber and sort based on the same column

```python
df.groupby(['Country', 'Population','Unemployment rate'])['Views_per_subscriber'].mean().reset_index().sort_values(by = 'Views_per_subscriber', ascending = False)
```

```python
df['views_per_person'] = (df['video views']/df['Population'])*100
df.groupby(['Country','Unemployment rate'])['views_per_person'].sum().reset_index().sort_values(by = 'views_per_person', ascending = False)
```
